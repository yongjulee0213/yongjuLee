```python
import requests
from bs4 import BeautifulSoup
```

```python
review_data=[]
pagenumber=1
while len(review_data)<=1000:
  URL=f"https://movie.naver.com/movie/point/af/list.naver?&page={pagenumber}"
  response=requests.get(URL)#URL얻어오기
  soup=BeautifulSoup(response.text, 'html.parser')#response.txt로 저장 html 파서 이용하기

  page_one=soup.find_all(name='td',attrs={'class':'title'}) #한칸씩 나오는거 확인

  for review in page_one:
    sentence=review.find(name='a',attrs={'class':'report'}).get("onclick").split("', '")[2]
    if sentence !="":#리뷰가 있어야 동작
      movie=review.find(name='a',attrs={'class':'movie color_b'}).string
      score=int(review.find(name='em').string)
      review_data.append([movie,sentence,score])
  pagenumber+1
```

크롤링 소요 시간 : 1m 7s

```python
print(len(review_data)) #데이터 수 확인
```

```python
#csv파일 
import csv

with open("samples.csv",'w') as fd:
  writer=csv.writer(fd,delimiter=',')
  writer.writerow(['movie', 'sentence','score'])#csv는 첫줄을 열이름으로 몰아붙이니까 미리 써놓기
  writer.writerows(review_data)
```
