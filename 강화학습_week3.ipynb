```python
#매직 명령어 사용 : 디렉토리이동, 파일 일람 표시등의 유틸리티 기능제공
#운영체제의 기능 끌어다가 사용
%pwd
%time sum(range(10000)) #time 뒤 동작이 실행시간 측정
%timeit sum(range(10000))#여러번 시행한 결과의 측정값 요약
%timeit -n 2000 -r 5 sum(range(10000)) #루프수 : 2000회, 시행횟수: 5
```

```python
%%timeit -n 1000 -r 3 #코드 셀 전체에 대해 러닝타임 특정

for i in range(1000):
    i*2
```
