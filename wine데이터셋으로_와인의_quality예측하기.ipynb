```python
bins=(0,5,10)#5를기준으로 작으면, bad, 크면 good
group_names=['bad', 'good']#정답으로 나올 것
wine['quality']=pd.cut(wine['quality'],bins=bins, labels=group_names)#quality항목 추가
```

```python
#set parameter
X=wine.drop('quality', axis=1)
Y=wine['quality']
```

```python
#데이터 나누기 cv
X_train, X_val, Y_train, Y_val=train_test_split(X, Y, test_size=0.1111, random_state=42, shuffle=False)
```

```python
def check_missing_col(dataframe):
    missing_col = []
    for col in dataframe.columns:
        missing_values = sum(dataframe[col].isna())
        is_missing = True if missing_values >= 1 else False
        if is_missing:
            print(f'결측치가 있는 컬럼은: {col} 입니다')
            print(f'해당 컬럼에 총 {missing_values} 개의 결측치가 존재합니다.')
            missing_col.append([col, dataframe[col].dtype])
    if missing_col == []:
        print('결측치가 존재하지 않습니다')
    return missing_col

missing_col = check_missing_col(X_train)
missing_col = check_missing_col(X_val)
```

```python
class LogisticRegression:
  def sigmoid(self,z):#sigmoid 1/(1+exp(-z))
    return 1/(1+numpy.exp(-z))

  def loss(self, thetas, X, Y):#crossentropy function 평균(ylog(y')+(1-y)log(1-y'))
    #y_hat=h(x)=g(z)
    z=numpy.dot(X, thetas)
    return numpy.mean(Y*numpy.log(self.sigmoid(z))+(1-Y)*numpy.log(1-self.sigmoid(z)))

  def train(self, X, Y,lr=0.01, max_iter=300):
    loss=[]
    thetas=numpy.random.rand(X_train.shape[1])#랜덤으로 쎄타설정
    m=X_train.shape[0]

    for epoch in range(max_iter):#반복하면서 쎼타 업데이트 loss저장
      y_pred=self.sigmoid(numpy.dot(X, thetas))#시그모이드에 넣으면 예측값구하는거
      thetas=thetas-(lr*numpy.dot(numpy.transpose(X), y_pred-Y)/m)#쎄타업데이트!!!
      loss.append(self.loss(thetas, X, Y))#loss 구해서 저장해놓기
    #이걸해야predict에서 구할수있음
    self.thetas=thetas
    self.loss=loss

  def predict(self, X):#예측값
    z=numpy.dot(X, self.thetas)
    return self.sigmoid(z)
```

```python
model1=LogisticRegression()
model1.train(X_train, Y_train)#훈련
```

```python
predict=model1.predict(X_val)
```
